```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import *
import sys
from pylab import *
```

```python
import pandas as pd
pd.set_option('display.max_rows', 600)
pd.set_option('display.max_columns', 600)
pd.set_option('display.width', 1000)
```

```python
df  = pd.read_csv ('adult_data.csv')
```

```python
df.head()
```

```python
df_marital = df.groupby('marital_s').aggregate(np.average)
df_marital
```

```python
sns.set(style="whitegrid")
tips = sns.load_dataset("tips")
ax =plt.xticks(rotation=90)
ax = sns.barplot(x=df.marital_s, y=df.income3, orient= 'v', data=df)
```

```python
sns.set(style="whitegrid")
tips = sns.load_dataset("tips")
f, axes = plt.subplots(3, 1, sharey=True, sharex=True, figsize=(6, 16))
#set(xlim=(0, 60)
ax= plt.xticks(rotation=90)
ax= sns.barplot(y=df.employer, x=df.income3, orient= 'h', data=df, ax=axes[0])
ax= sns.barplot(y=df.race, x=df.income3,  orient= 'h', data=df, ax=axes[1])
ax= sns.barplot(y=df.marital_s, x=df.income3,  orient= 'h', data=df, ax=axes[2])
```

```python
sns.set(style="whitegrid")
tips = sns.load_dataset("tips")
ax1 = plt.xticks(rotation=90)
ax1 = sns.barplot(x=df_marital.index, y=df_marital['income3'], orient= 'v', data=tips)
```

```python
df_tmp = df.groupby('race').aggregate(np.average)
sns.set(style="whitegrid")
tips = sns.load_dataset("tips")
ax2 = plt.xticks(rotation=90)
ax2 = sns.barplot(x=df_tmp.index, y=df_tmp['income3'], orient= 'v', data=tips)
```

```python
df_tmp = df.groupby('employer').aggregate(np.average)
sns.set(style="whitegrid")
tips = sns.load_dataset("tips")
ax3 = plt.xticks(rotation=90)
ax3 = sns.barplot(y=df_tmp['income3'], x=df_tmp.index, orient= 'v', data=tips)
```

```python

fig = plt.figure(figsize=(13, 8))
df_tmp = df.groupby('occupation').aggregate(np.average)
sns.set(style="whitegrid")
ax = plt.xticks(rotation=90)
tips = sns.load_dataset("tips")
ax = sns.barplot(y=df_tmp['income3'], x=df_tmp.index, orient= 'v', data=tips)
```

```python
df_tmp = df.groupby('education').aggregate(np.average)
sns.set(style="whitegrid")
fig = plt.figure(figsize=(13, 8))
ax2 = plt.xticks(rotation=90)
tips = sns.load_dataset("tips")
ax = sns.barplot(y=df_tmp['income3'], x=df_tmp.index, orient= 'v', data=tips)
```

```python
df_tmp = df.groupby('gender').aggregate(np.average)
sns.set(style="whitegrid")
tips = sns.load_dataset("tips")
ax = sns.barplot(y=df_tmp['income3'], x=df_tmp.index, orient= 'v', data=tips)
```

```python
df_tmp = df.groupby('country').aggregate(np.average)
sns.set(style="whitegrid")
fig = plt.figure(figsize=(10, 16))
tips = sns.load_dataset("tips")
ax = sns.barplot(x=df_tmp['income3'], y=df_tmp.index, orient= 'h', data=tips)
```

```python
sns.set()
fig = plt.figure(figsize=(6, 8))
ax3 = plt.subplot2grid((2, 2), (1, 0), colspan=2)
ax3 = plt.xticks(rotation=0)
ax3 = sns.boxplot(x="income2", y="age", data=df)
ax3 = plt.title('Income per Age')
```

```python
sns.set()
fig = plt.figure(figsize=(10, 16))
ax3 = plt.subplot2grid((2, 2), (1, 0), colspan=2)
ax3 = plt.xticks(rotation=0)
ax3 = sns.violinplot(x="income2", y="age", data=df)
ax3 = plt.title('Income per Age')
```
